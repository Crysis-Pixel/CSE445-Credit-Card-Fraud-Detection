```python
# Separate features and target
non_fraud_data = new_df[new_df['Class'] == 0]
non_fraud_data_reduced = non_fraud_data.sample(n=2000, random_state=42)
fraud_data = new_df[new_df['Class'] == 1]
X = fraud_data.drop('Class', axis=1)  # Features
y = fraud_data['Class']  # Target variable
new_df = pd.concat([non_fraud_data_reduced, fraud_data])
# Separate features and target
X = new_df.drop('Class', axis=1)  # Features (drop the target column 'Class')
y = new_df['Class']  # Target variable

# Apply SMOTE to balance the classes
smote = SMOTE(sampling_strategy= 0.5, random_state=42)  # 'auto' means balancing to the majority class
X_resampled, y_resampled = smote.fit_resample(X, y)

# Create a balanced DataFrame with resampled data
balanced_data = pd.DataFrame(X_resampled, columns=X.columns)
balanced_data['Class'] = y_resampled

# Check class distribution
print("Class distribution after SMOTE oversampling:")
print(balanced_data['Class'].value_counts())
```

```python
x = augmented_data.drop('Class', axis=1)
y = augmented_data['Class']
```

```python
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
#addded class_weight='balanced' to avoid zero division
```

```python
classifier = {
    "Logistic Regression": LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42),
    "Decision Tree": DecisionTreeClassifier(class_weight='balanced', random_state=42),
    "Random Forest": RandomForestClassifier(class_weight='balanced', random_state=42),
    "SVM": SVC(class_weight='balanced', random_state=42)
    
}
```

```python
combined_cm_dict = {name: np.zeros((2, 2), dtype=int) for name in classifier.keys()} 

# Initialize dictionaries to store metrics for each algorithm
accuracy_dict = {name: [] for name in classifier.keys()}
precision_dict = {name: [] for name in classifier.keys()}
recall_dict = {name: [] for name in classifier.keys()}
f1_score_dict = {name: [] for name in classifier.keys()}

# Use StratifiedKFold on the training set only
for fold, (train_index, test_index) in enumerate(skf.split(x, y)):
    print(f"\nFold {fold + 1}")
    
    # Split the data for this fold
    x_fold_train = x.iloc[train_index]
    x_fold_test = x.iloc[test_index]
    y_fold_train = y.iloc[train_index]
    y_fold_test = y.iloc[test_index]

    # Train the classifier for each model
    for name, clf in classifier.items():
        print(f"\n========== {name} - Fold {fold + 1} ==========")
        clf.fit(x_fold_train, y_fold_train)
        y_pred = clf.predict(x_fold_test)

        # Calculate metrics for this fold
        acc = accuracy_score(y_fold_test, y_pred)
        prec = precision_score(y_fold_test, y_pred, zero_division=0)
        rec = recall_score(y_fold_test, y_pred, zero_division=0)
        f1 = f1_score(y_fold_test, y_pred, zero_division=0)

        # Append metrics to the respective algorithm's list
        accuracy_dict[name].append(acc)
        precision_dict[name].append(prec)
        recall_dict[name].append(rec)
        f1_score_dict[name].append(f1)

        # Print individual metrics for the current fold
        print(f"  Accuracy: {acc:.4f}")
        print(f"  Precision: {prec:.4f}")
        print(f"  Recall: {rec:.4f}")
        print(f"  F1 Score: {f1:.4f}")

        # Update combined confusion matrix
        cm = confusion_matrix(y_fold_test, y_pred)
        combined_cm_dict[name] += cm  # Add fold's confusion matrix to combined matrix

        # Plot confusion matrix for this fold
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                    xticklabels=np.unique(y_fold_train), 
                    yticklabels=np.unique(y_fold_train))
        plt.title(f"{name} - Fold {fold + 1}")
        plt.xlabel("Predicted Labels")
        plt.ylabel("Actual Labels")
        plt.show()

# Plot combined confusion matrices for each algorithm
for name, combined_cm in combined_cm_dict.items():
    plt.figure(figsize=(8, 8))
    sns.heatmap(combined_cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=["Non-Fraud", "Fraud"], 
                yticklabels=["Non-Fraud", "Fraud"],
                annot_kws={"size": 30}) 
    plt.title(f"{name} - Combined Confusion Matrix", fontsize=18)  
    plt.xlabel("Predicted Labels", fontsize=16)  
    plt.ylabel("Actual Labels", fontsize=16)  
    plt.xticks(fontsize=16)  
    plt.yticks(fontsize=16)  
    plt.show()
    
# After all folds are processed, compute and print the average of each metric per algorithm
for name in classifier.keys():
    avg_accuracy = sum(accuracy_dict[name]) / len(accuracy_dict[name])
    avg_precision = sum(precision_dict[name]) / len(precision_dict[name])
    avg_recall = sum(recall_dict[name]) / len(recall_dict[name])
    avg_f1_score = sum(f1_score_dict[name]) / len(f1_score_dict[name])

    print(f"\n========== Average Metrics for {name} ==========")
    print(f"  Average Accuracy: {avg_accuracy:.4f}")
    print(f"  Average Precision: {avg_precision:.4f}")
    print(f"  Average Recall: {avg_recall:.4f}")
    print(f"  Average F1 Score: {avg_f1_score:.4f}")
```

```python
# Prepare the data for plotting the average metrics for all algorithms
algorithms = list(classifier.keys())
avg_accuracy = [sum(accuracy_dict[name]) / len(accuracy_dict[name]) for name in algorithms]
avg_precision = [sum(precision_dict[name]) / len(precision_dict[name]) for name in algorithms]
avg_recall = [sum(recall_dict[name]) / len(recall_dict[name]) for name in algorithms]
avg_f1_score = [sum(f1_score_dict[name]) / len(f1_score_dict[name]) for name in algorithms]

# Create a grouped bar chart for the average metrics
x = np.arange(len(algorithms)) 
width = 0.2 

fig, ax = plt.subplots(figsize=(14, 6))

# Plot bars for each metric with new color scheme
rects1 = ax.bar(x - 1.5*width, avg_accuracy, width, label='Accuracy', color='#4271C5') 
rects2 = ax.bar(x - 0.5*width, avg_precision, width, label='Precision', color='#A5A5A5')  
rects3 = ax.bar(x + 0.5*width, avg_recall, width, label='Recall', color='#FFC000') 
rects4 = ax.bar(x + 1.5*width, avg_f1_score, width, label='F1 Score', color='#ED7D2D')  

# Add some text for labels, title, and custom x-axis tick labels
ax.set_xlabel('Machine Learning Method', fontsize=14)
ax.set_ylabel('Mean Score Across (k=10) Folds', fontsize=14)
ax.set_title('K-Fold (k=10) Cross-Validation Results\nMultiple Machine Learning Methods and Metrics', fontsize=16)
ax.set_xticks(x)
ax.set_xticklabels(algorithms, fontsize=12)
ax.legend()

# Annotate the bars with the values
def annotate_bars(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.4f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=12)

annotate_bars(rects1)
annotate_bars(rects2)
annotate_bars(rects3)
annotate_bars(rects4)

# Display the plot
plt.tight_layout()
plt.show()
```
